# 输入和输出

## 输出

```python
# 输出字符串
print("hello world")
# 输出多个字符串
print("hello world", "hello world", "hello world")
# 输出整数
print(100)
# 输出多个整数
print(100, 100, 100)
# 输出计算结果
print(100+100)
# 输出字符串加计算结果
print("100+100 =", 100+100)
```

## 输入

```python
# 字符串输入
name = input("请输入你的名字：")
print("hello", name)
print(type(name))
# 强制转换为数字
age = int(input("请输入你的年龄："))
print("age:", age)
print(type(age))
```

# 数据类型和变量

## 整数

```python
a = 10
print(a)
print(type(a))
```

## 浮点数

```python
a = 1.23
print(a)
print(type(a))
```

## 字符串

```python
# 普通字符串
a = 'hello world'
print(a)
print(type(a))
# 换行符
a = 'hello\nworld'
print(a)
# 缩进符
a = 'hello\tworld'
print(a)
# 原始字符串->忽略所有转义符
a = r'hello\nworld'
print(a)
```

## 布尔值

```python
# 布尔值只有两种，True和False
print(True)
print(False)
print(type(True))
print(type(False))

# 逻辑运算符
print("True and True =", True and True)
print("True and False =", True and False)
print("False and False =", False and False)
print("True or True =", True or True)
print("True or False =", True or False)
print("False or False =", False or False)
print("not True =", not True)
print("not False =", not False)
```
